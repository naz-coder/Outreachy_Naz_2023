# 2.0 BIOCLIMATIC AND ECOLOGICAL ZONES

### 2.1 Climate
The Federal Capital Territory (FCT) encounters distinctive climate patterns. In the dry seasons, the FCT witnesses high temperatures with minimal cloud cover. Positioned in the central belt of Nigeria, the FCT enjoys a milder climate, avoiding the extreme conditions found in the northern and southern regions of the country. It experiences two prominent seasons: the wet season and the dry season. Notably, there can be a substantial temperature variation of up to 17C in a single day.

During the rainy season, the FCT experiences lower maximum temperatures, primarily because of the thick cloud cover. The diurnal annual temperature range is also considerably reduced, sometimes not exceeding 7C, particularly in July and August, when the temperature fluctuates between 30.4C and 35.1C.

## 5.0 Forest
### 5.1 Planted forest
According to the U.N. FAO, 9.9% or about 9,041,000 ha of Nigeria is forested, according to FAO. Nigeria had 382,000 ha of planted forest. Change in Forest Cover: Between 1990 and 2010, Nigeria lost an average of 409,650 ha or 2.38% per year. In total, between 1990 and 2010, Nigeria lost 47.5% of its forest cover or around 8,193,000 ha. Nigeria&apos;s forests contain 1,085 million metric tons of carbon in living forest biomass. Biodiversity and Protected Areas: Nigeria has some 1417 known species of amphibians, birds, mammals and reptiles according to figures from the World Conservation Monitoring Centre. Of these, 1.2% are endemic, meaning they exist in no other country, and 3.5% are threatened. Nigeria is home to at least 4715 species of vascular plants, of which 4.3% are endemic according to Mongapay(2011).
